```python
heights = tensor([16., 20., 22., 24., 40., 52., 53., 56., 63., 82.])
flowers = tensor([2., 2., 3., 2., 4., 5., 6., 5., 6., 8.])
x = heights
y = flowers
```

```python
def predict(theta0, theta1, x):
  return theta0 + theta1 * x
```

```python
def loss(theta0, theta1):
  return ((predict(theta0, theta1, x) - y) ** 2).mean()
```

```python
lr = 1e-4
```

```python
theta0 = tensor([np.random.normal()]).requires_grad_()
theta1 = tensor([np.random.normal()]).requires_grad_()
```

```python
L = loss(theta0, theta1)
L
```
